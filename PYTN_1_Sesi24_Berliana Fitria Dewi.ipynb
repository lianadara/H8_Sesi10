## Area Pots (Series/Dataframe) <a id="12"></a>

**What is a line plot and why use it?**

An Area chart or area plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields. Use line plot when you have a continuous data set. These are best suited for trend-based visualizations of data over a period of time.

**Questions:**

Berapa jumlah kasus kriminalitas yang terjadi di tiap tahunnya?

```python
crime_years = df[(df['year'] >= 2008) & (df['year'] <= 2016)]

# Group by year and major_category, then sum the values
grouped_data = crime_years.groupby(['year', 'major_category'])['value'].sum().unstack().fillna(0)

# sorted the column
sorted_columns = grouped_data.sum().sort_values(ascending=True).index
grouped_data = grouped_data[sorted_columns]

# Custom color
palette = sns.color_palette("pastel")

# Membuat area chart
plt.figure(figsize=(12, 8))
ax = grouped_data.plot(kind='area', stacked=True, alpha=0.8, color=palette)

# Menambahkan label angka di atas area chart
for year in grouped_data.index:
    total_value = grouped_data.loc[year].sum()
    plt.text(year, total_value + 500, f'{total_value/1000:.0f}k', ha='center', va='bottom', fontsize=8, color='black')

plt.legend(title='Major Crime Categories', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Distribusi Kejahatan Berdasarkan Kategori Major per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Kasus')

# Graded-Funtion End
plt.show()
```

**Insight:**
Tingkat kriminalitas sangat tinggi di tahun 2008 dan sempat menurun di tahun 2014 kemudian kembali meningkat di tahun 2016. Peningkatan dari tahun 2014 ke 2016 sangat tajam sehingga harus ada analisis lebih lanjut untuk memahami penyebab peningkatan tajam dalam tingkat kriminalitas dari tahun 2014 ke 2016. Selain itu, juga bisa dilakukan analisis penyebab keberhasilan penurunan kasus kriminalitas di tahun 2014 sehingga dapat diterapkan untuk tahun-tahun berikutnya.

## Histogram

**Questions:**

Berapa jumlah kasus kriminalitas yang terjadi di tiap bulan pada tahun 2016?

```python
# Mengelompokkan data dan menjumlahkan nilai (value) per bulan
total_crimes_by_month = df[df['year']==2016].groupby('month')['value'].sum()

# Membuat histogram dengan palet warna merah
plt.figure(figsize=(10, 6))
plt.hist(total_crimes_by_month.index, bins=range(1, 14), weights=total_crimes_by_month, edgecolor='black', align='left')

# Menambahkan label angka di atas setiap bar dengan format yang disingkat
for i, value in enumerate(total_crimes_by_month):
    plt.text(i + 1, value + 500, f'{value/1000:.0f}k', ha='center', va='bottom', fontsize=10, color='black')

plt.title('Histogram Total Kejahatan per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Kejahatan')
plt.xticks(range(1, 13))
plt.show()
```

**Insight:**

Dari histogram tersebut, terlihat bahwa jumlah kejahatan paling banyak terjadi di bulan Juli 2016 dan peningkatan paling tajam terjadi di bulan Mei. Tingkat kejahatan cenderung meningkat di bulan-bulan terakhir sehingga keamanan dapat diperketat di bulan tersebut.

## Bar Charts (Dataframe) <a id="10"></a>

A bar plot is a way of representing data where the *length* of the bars represents the magnitude/size of the feature/variable. Bar graphs usually represent numerical and categorical variables grouped in intervals. 

To create a bar plot, we can pass one of two arguments via `kind` parameter in `plot()`:

* `kind=bar` creates a *vertical* bar plot
* `kind=barh` creates a *horizontal* bar plot

**Question:**

Berapa jumlah kejahatan yang terjadi di tiap daerah?

```python
# Menghitung total kejahatan per borough
total_crimes_by_borough = df.groupby('borough')['value'].sum().sort_values(ascending=False)

# Membuat bar chart dengan palet warna pastel
plt.figure(figsize=(12, 8))
ax = sns.barplot(x=total_crimes_by_borough.values, y=total_crimes_by_borough.index, palette='Reds_r')
plt.title('Total Kejahatan per Daerah')
plt.xlabel('Jumlah Kasus Kriminalitas')
plt.ylabel('Daerah')

# Menambahkan angka di samping bar
for i, value in enumerate(total_crimes_by_borough.values):
    ax.text(value + 1000, i, f'{value/1000:.0f}k', ha='left', va='center', fontsize=10, color='black')

plt.show()
```

**Insight:**

Wilayah Westminster memiliki jumlah kasus yang paling banyak hingga melebihi 450.000 kasus, cukup jauh dari jumlah kejahatan di daerah-daerah lainnya. Hal ini perlu dilakukan penyelidikan dan keamanan yang lebih tinggi di daerah tersebut atau mencontoh daerah City of London yang memiliki jumlah kriminalitas terendah.

## Pie Charts <a id="6"></a>

A `pie chart` is a circualr graphic that displays numeric proportions by dividing a circle (or pie) into proportional slices. You are most likely already familiar with pie charts as it is widely used in business and media. We can create pie charts in Matplotlib by passing in the `kind=pie` keyword.

**Question:**

Apa kasus kejahatan yang paling sering terjadi?

```python
# Menghitung jumlah kejahatan per kategori utama
total_crimes_by_major_category = df.groupby('major_category')['value'].sum()

# Menentukan batas persentase kecil yang akan digabungkan
threshold = 5.0  # Misalnya, persentase kurang dari 5% akan digabungkan ke dalam kategori "Others"

# Membuat dataframe baru dengan kategori "Others"
small_categories = total_crimes_by_major_category[total_crimes_by_major_category / total_crimes_by_major_category.sum() * 100 < threshold]
df['major_category'] = df['major_category'].replace(small_categories.index, 'Others')

# Menghitung kembali jumlah kejahatan per kategori utama
total_crimes_by_major_category = df.groupby('major_category')['value'].sum()

# Membuat pie chart dengan hanya menampilkan 5 kategori teratas
top_categories = total_crimes_by_major_category.sort_values(ascending=False).head(5)
top_categories['Others'] = total_crimes_by_major_category[~total_crimes_by_major_category.index.isin(top_categories.index)].sum()

# Membuat pie chart
plt.figure(figsize=(10, 8))
plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title('Proporsi Jumlah Kejahatan Berdasarkan Major Category')
plt.show()
```

**Insight:**
Kasus kejahatan yang paling banyak terjadi adalah pencurian sehingga penting untuk memprioritaskan pencegahan di kategori ini. Pencegahan juga dapat dilakukan dengan memberikan himbauan kepada masyarakat untuk memperketat tindakan pencegahan, atau juga dapat dilakukan dengan meningkatkan hukuman pada pencuri.

## Box Plots <a id="8"></a>

A `box plot` is a way of statistically representing the *distribution* of the data through five main dimensions: 

- **Minimun:** Smallest number in the dataset.
- **First quartile:** Middle number between the `minimum` and the `median`.
- **Second quartile (Median):** Middle number of the (sorted) dataset.
- **Third quartile:** Middle number between `median` and `maximum`.
- **Maximum:** Highest number in the dataset.

## Scatter Plots <a id="10"></a>

A `scatter plot` (2D) is a useful method of comparing variables against each other. `Scatter` plots look similar to `line plots` in that they both map independent and dependent variables on a 2D graph. While the datapoints are connected together by a line in a line plot, they are not connected in a scatter plot. The data in a scatter plot is considered to express a trend. With further analysis using tools like regression, we can mathematically calculate this relationship and use it to predict trends outside the dataset.

**Question:**

Bagaimana korelasi jumlah kejahatan antar majorcategory di tahun 2016?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Menggabungkan data berdasarkan tahun dan major category
combined_data = df[df['year']==2016].groupby(['month', 'major_category']).agg({'value': 'sum'}).reset_index()

# Membuat scatter plot
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(x='month', y='value', hue='major_category', data=combined_data, s=200)
plt.title('Scatter Plot Korelasi Jumlah Kejahatan antar Major Category per Bulan di Tahun 2016')
scatter.legend(title='Kategori Utama', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('Bulan')
plt.ylabel('Jumlah Kejahatan')
plt.show()
```

**Insight:**

Berdasarkan plot di tahun 2016 tersebut, terlihat bahwa kasus kejahatan yang terjadi cenderung memiliki korelasi di antar kategorinya, salah satunya yaitu terdapat korelasi positif antara jumlah kejahatan kekerasan terhadap jumlah pencurian. Hal ini berarti bahwa ketika jumlah kekerasan meningkat maka jumlah pencurian juga cenderung meningkat. Selain itu, terdapat korelasi negatif yang lemah antara jumlah kekerasan dengan jumlah kasus narkoba. Hal ini berarti ketika kasus narkoba meningkat, maka jumlah kekerasan cenderung menurun.

## Folium

Folium is a powerful Python library that helps you create several types of Leaflet maps. The fact that the Folium results are interactive makes this library very useful for dashboard building.

From the official Folium documentation page:

> Folium builds on the data wrangling strengths of the Python ecosystem and the mapping strengths of the Leaflet.js library. Manipulate your data in Python, then visualize it in on a Leaflet map via Folium.

> Folium makes it easy to visualize data that's been manipulated in Python on an interactive Leaflet map. It enables both the binding of data to a map for choropleth visualizations as well as passing Vincent/Vega visualizations as markers on the map.

> The library has a number of built-in tilesets from OpenStreetMap, Mapbox, and Stamen, and supports custom tilesets with Mapbox or Cloudmade API keys. Folium supports both GeoJSON and TopoJSON overlays, as well as the binding of data to those overlays to create choropleth maps with color-brewer color schemes.

```python
pip install folium
```

```python
import folium

print('Folium installed and imported!')
```

```python
import folium
import geopandas as gpd
from folium import Choropleth
import json

# Assuming df_crimes_val is your DataFrame
borough_crimes = df.groupby('borough')['value'].sum().reset_index()
boroughs_in_df = borough_crimes['borough'].unique()

# Load GeoJSON data (download from: https://github.com/radoi90/housequest-data/tree/master)
# Load GeoJSON data using geopandas
geojson_url = "https://raw.githubusercontent.com/radoi90/housequest-data/master/london_boroughs.geojson"
boroughs_data = gpd.read_file(geojson_url)
geojson_path = 'london_boroughs.geojson'
with open(geojson_path, 'r') as geojson_file:
    geojson_data = json.load(geojson_file)

# Filter GeoJSON data to include only the relevant boroughs
filtered_geojson = {
    "type": geojson_data["type"],
    "features": [feature for feature in geojson_data["features"] if feature["properties"]["name"] in boroughs_in_df]
}

# Create a Folium map centered around London
london_map = folium.Map(location=[51.509865, -0.118092], zoom_start=10)

# Create a choropleth map layer
Choropleth(
    geo_data=filtered_geojson,
    data=borough_crimes,
    columns=['borough', 'value'],
    key_on='feature.properties.name',
    fill_color='YlGnBu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Number of Crimes',
).add_to(london_map)

# Add borough labels
for feature in filtered_geojson['features']:
    name = feature['properties']['name']
    coords = feature['geometry']['coordinates'][0][0]  # Use the coordinates of one of the borough's points
    marker = folium.Marker(
        location=[coords[1], coords[0]],
        tooltip=name
    )
    marker.add_to(london_map)


# Display the map in the notebook
london_map
```
